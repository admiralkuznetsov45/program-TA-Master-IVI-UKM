```python
import pandas as pd
```

```python

nama_file_csv = 'full_test_split.csv'  # Ganti dengan nama file CSV Anda

try:
    # Baca file CSV
    df = pd.read_csv(nama_file_csv)

    # Tampilkan isi DataFrame (tabel)
    print(df)

    # Opsi tambahan: Tampilkan beberapa baris pertama
    # print(df.head())

    #Opsi tambahan : Tampilkan info dari dataframe
    print(df.info())

except FileNotFoundError:
    print(f"Error: File '{nama_file_csv}' tidak ditemukan.")
except Exception as e:
    print(f"Terjadi kesalahan: {e}")
```

```python

def klasifikasi_depresi(nama_file_csv):
    """
    Membaca file CSV, mengklasifikasikan depresi/non-depresi berdasarkan PHQ_Score, dan menampilkan hasilnya.

    Args:
        nama_file_csv (str): Nama file CSV input.
    """
    try:
        # Baca file CSV
        df = pd.read_csv(nama_file_csv)

        # Membuat kolom 'Klasifikasi' berdasarkan PHQ_Score
        df['Klasifikasi'] = df['PHQ_Score'].apply(lambda x: 'Depresi' if x >= 10 else 'Non-Depresi')

        # Tampilkan DataFrame dengan kolom 'Klasifikasi'
        print(df)

    except FileNotFoundError:
        print(f"Error: File '{nama_file_csv}' tidak ditemukan.")
    except Exception as e:
        print(f"Terjadi kesalahan: {e}")

def klasifikasi_dan_simpan_csv(nama_file_input, nama_file_output):
    """
    Membaca file CSV, mengklasifikasikan depresi/non-depresi, dan menyimpan ke file CSV baru.

    Args:
        nama_file_input (str): Nama file CSV input.
        nama_file_output (str): Nama file CSV output.
    """
    try:
        # Baca file CSV
        df = pd.read_csv(nama_file_input)

        # Klasifikasikan depresi/non-depresi
        df['Klasifikasi'] = df['PHQ_Score'].apply(lambda x: 'Depresi' if x >= 10 else 'Non-Depresi')

        print(df)

        # Simpan DataFrame ke file CSV baru
        df.to_csv(nama_file_output, index=False)  # index=False agar indeks tidak ikut disimpan

        print(f"File CSV '{nama_file_input}' berhasil diolah dan disimpan sebagai '{nama_file_output}'.")

    except FileNotFoundError:
        print(f"Error: File '{nama_file_input}' tidak ditemukan.")
    except Exception as e:
        print(f"Terjadi kesalahan: {e}")

# Contoh penggunaan
# nama_file_csv = 'full_test_split.csv'  # Ganti dengan nama file CSV Anda
# klasifikasi_depresi(nama_file_csv)

nama_file_input = 'full_test_split.csv'  # Ganti dengan nama file CSV input Anda
nama_file_output = 'hasil_klasifikasi.csv'  # Ganti dengan nama file CSV output yang diinginkan
klasifikasi_dan_simpan_csv(nama_file_input, nama_file_output)
```

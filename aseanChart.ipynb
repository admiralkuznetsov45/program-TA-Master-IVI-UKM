```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# --- Mengimpor data dari Excel ---

# Ganti 'path/to/your/file.xlsx' dengan path ke file Excel Anda
df = pd.read_excel('\program-TA-master\ASEAN-Elderly.xlsx')  

print(df)
```

```python
# --- Mengambil data yang dibutuhkan ---

categories = df['Negara'].tolist()
age_60_64_2000 = df['60-64 (2000)'].to_numpy()
age_65_2000 = df['65+ (2000)'].to_numpy()
age_60_64_2022 = df['60-64 (2022)'].to_numpy()
age_65_2022 = df['65+ (2022)'].to_numpy()

# --- Membuat stacked bar chart ---

plt.figure(figsize=(15, 6))
width = 0.35  # Lebar setiap batang

# Batang untuk tahun 2000
plt.bar(np.arange(len(categories)) - width/2, age_60_64_2000, width, color='#0099cb', label='60-64 (2000)')
plt.bar(np.arange(len(categories)) - width/2, age_65_2000, width, bottom=age_60_64_2000, color='#a3c6e3', label='≥ 65 (2000)')

# Batang untuk tahun 2022
plt.bar(np.arange(len(categories)) + width/2, age_60_64_2022, width, color='#0070a3', label='60-64 (2022)')
plt.bar(np.arange(len(categories)) + width/2, age_65_2022, width, bottom=age_60_64_2022, color='#70a3c6', label='≥ 65 (2022)')

# Menambahkan label persentase
for i, cat in enumerate(categories):
    plt.text(i - width/2, age_60_64_2000[i]/2, f"{age_60_64_2000[i]:.1f}%", ha='center', va='center', color='black')
    plt.text(i - width/2, age_60_64_2000[i]+age_65_2000[i]/2, f"{age_65_2000[i]:.1f}%", ha='center', va='center', color='black')
    plt.text(i + width/2, age_60_64_2022[i]/2, f"{age_60_64_2022[i]:.1f}%", ha='center', va='center', color='black')
    plt.text(i + width/2, age_60_64_2022[i]+age_65_2022[i]/2, f"{age_65_2022[i]:.1f}%", ha='center', va='center', color='black')

# Mengatur label sumbu dan judul
plt.xlabel("Country")
plt.ylabel("Percent")
plt.title("Elderly Statistics in ASEAN (Between 2000-2022)")
plt.xticks(np.arange(len(categories)), categories)
plt.legend()
plt.tight_layout()
plt.show()
```
